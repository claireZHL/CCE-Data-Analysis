```python
import numpy as np
import pandas as pd
```

```python
# Load unlabeled Survey Data
survey = pd.read_csv('cce_survey_data.csv').dropna(subset=['zipcode'])

# Clean unlabeled Survey data
survey['zipcode'] = survey['zipcode'].astype(int)
survey
```

```python
# Load 6 clusters zipcode data
zipcode = pd.read_csv('labeled SD zipcode 6 clusters.csv')
zipcode
```

```python
# Assign Label to Survey data
def assign_label(value):
    for i in range(len(zipcode)):
        if value == zipcode['Zipcode'][i]:
            return zipcode['label'][i]
    return 'NaN'
        
survey['label'] = survey['zipcode'].apply(assign_label)
survey
```

```python
# Export Labeled Survey as csv file
csv_filename = 'labeled_cce_survey_data.csv'
survey.to_csv(csv_filename, index=False)

# Export Labeled Survey as excel file
excel_filename = 'labeled_cce_survey_data.xlsx'
survey.to_excel(excel_filename, index=False)
```
